```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
df.info()
```

```python
correlation = df[['Conversion_Rate',
                  'Acquisition_Cost',
                  'ROI',
                  'Clicks',
                  'Impressions',
                  'Engagement_Score',
                  'CTR',
                  'Profit_Index']].corr()

correlation
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.heatmap(correlation, annot=True)

plt.title("Marketing Metrics Correlation Heatmap")

plt.show()
```

```python
plt.savefig(r"C:\Users\rajan\OneDrive\Desktop\DATA ANALYST\DATA ANALYST PROJECT\Marketing_Campaign_Analysis\Screenshot")
```

```python
channel_roi = df.groupby('Channel_Used')['ROI'].mean()

channel_roi
```

```python
channel_roi.plot(kind='bar')

plt.title("Average ROI by Marketing Channel")

plt.xlabel("Marketing Channel")

plt.ylabel("Average ROI")

plt.ylim(channel_roi.min() - 0.02,
         channel_roi.max() + 0.02)

plt.show()
```

```python
segment_profit = df.groupby('Customer_Segment')['Profit_Index'].mean()

segment_profit
```

```python
segment_profit.plot(kind='bar')

plt.title("Average Profit by Customer Segment")

plt.ylim(segment_profit.min() - 100,
         segment_profit.max() + 100)

plt.show()
```

```python
monthly_roi = df.groupby(df['Date'].dt.month)['ROI'].mean()

monthly_roi
```

```python
monthly_roi.plot(marker='o')

plt.title("Monthly ROI Trend")

plt.xlabel("Month")

plt.ylabel("Average ROI")

plt.show()
```
